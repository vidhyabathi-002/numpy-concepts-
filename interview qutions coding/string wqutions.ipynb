count vowel and consonent 

```python
s = "rio is raj and ramya"

v = 0
c = 0

for i in s:
    if i in "AEIOUaeiou":
        v += 1
    elif ('a' <= i <= 'z') or ('A' <= i <= 'Z'):
        c += 1

print("Vowels =", v)
print("Consonants =", c)
```

```python
##Find Smallest Word in a Sentence

s= "rio is raj and ramya tk roieowkdnbemnwxk  wbdkhdui edk qwh y"
word=s.split()
print(word)
print(min(word))#####min() normally finds the smallest value alphabetically.
print ("Smallest word is :",min( word, key=len))    
#But key=len changes the comparison rule.
#Now Python checks the length of each word.
```
