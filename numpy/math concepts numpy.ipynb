```python
# Example array for all operations
arr = np.array([10, 20, 30, 40, 50])

# 1. max() - Find maximum value
print("1. max():")
print(f"   Array: {arr}")
print(f"   Maximum value: {np.max(arr)}")
print()

# 2. min() - Find minimum value
print("2. min():")
print(f"   Minimum value: {np.min(arr)}")
print()

# 3. mean() - Calculate average
print("3. mean():")
print(f"   Average (mean): {np.mean(arr)}")
print()

# 4. median() - Find middle value
print("4. median():")
print(f"   Median value: {np.median(arr)}")
print()

# 5. mod() - Modulo operation
print("5. mod():")
arr2 = np.array([12, 15, 20])
divisor = 5
print(f"   Array: {arr2}")
print(f"   Modulo {divisor}: {np.mod(arr2, divisor)}")
print()

# 6. variance()  - Measure of spread
print("6. variance():")
print(f"   Variance: {np.var(arr)}")
print()

# 7. std() - Standard deviation
print("7. std():")
print(f"   Standard Deviation: {np.std(arr)}")
```

### **Additional Array Operations:**

* **abs()** - Convert all negative values to positive
  - **syntax:** `np.abs(variable)`
  
* **size()** - Find the number of elements in array
  - **syntax:** `np.size(variable)`
  
* **sum()** - Calculate sum of all array elements
  - **syntax:** `np.sum(variable)`

```python
# Example array with negative values
arr_with_neg = np.array([-5, 10, -15, 20, -25])

# 1. abs() - Convert negative to positive
print("1. abs():")
print(f"   Original array: {arr_with_neg}")
print(f"   Absolute values: {np.abs(arr_with_neg)}")
print()

# 2. size() - Find number of elements
print("2. size():")
arr = np.array([10, 20, 30, 40, 50])
print(f"   Array: {arr}")
print(f"   Number of elements: {np.size(arr)}")
print()

# 3. sum() - Calculate sum of all elements
print("3. sum():")
print(f"   Array: {arr}")
print(f"   Sum of all elements: {np.sum(arr)}")
print(f"   Total: 10 + 20 + 30 + 40 + 50 = {np.sum(arr)}")
```

# Basic Arithmetic Operations with NumPy

NumPy supports element-wise arithmetic operations on arrays:
- **Addition (+)** - Add values
- **Subtraction (-)** - Subtract values
- **Multiplication (*)** - Multiply values
- **Division (/)** - Divide values
- **Floor Division (//)** - Divide and round down
- **Power (**)** - Raise to power
- **Modulo (%)** - Remainder after division

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])

print("Original array:", arr)
print()

# Addition
print("1. Addition (arr + 2):")
print(arr + 2)
print()

# Subtraction
print("2. Subtraction (arr - 2):")
print(arr - 2)
print()

# Multiplication
print("3. Multiplication (arr * 2):")
print(arr * 2)
print()

# Division
print("4. Division (arr / 2):")
print(arr / 2)
print()

# Floor Division
print("5. Floor Division (arr // 2):")
print(arr // 2)
print()

# Power
print("6. Power (arr ** 2):")
print(arr ** 2)
print()

# Modulo
print("7. Modulo (arr % 2):")
print(arr % 2)
print()

# Operations with two arrays
arr2 = np.array([10, 20, 30, 40, 50])
print("Array 1:", arr)
print("Array 2:", arr2)
print()
print("Element-wise addition (arr + arr2):")
print(arr + arr2)
```
